```python
def getExtremes(df, col_name, max_std = 3):
    mean = df[col_name].mean() 
    std_max = df[col_name].std() * max_std
    too_high = mean + std_max
    too_low = mean - std_max

    extreme = df[(df[col_name] > too_high) | (df[col_name] < too_low)]
    return extreme
```

```python
def createInteraction(df, test_df, interactions, new_col_name, conversion_dict = {}):
    if len(conversion_dict):
        if(df[interactions[0]].dtype.name=="object"):
            df[interactions[0]] = df[interactions[0]].map(conversion_dict)
            test_df[interactions[0]] = test_df[interactions[0]].map(conversion_dict)
        if(df[interactions[1]].dtype.name=="object"):
            test_df[interactions[1]] = test_df[interactions[1]].map(conversion_dict)
            df[interactions[1]] = df[interactions[1]].map(conversion_dict)            

    df[new_col_name]      = df[interactions[0]] * df[interactions[1]]
    test_df[new_col_name] = test_df[interactions[0]] * test_df[interactions[1]]
    
    if len(conversion_dict) == 3:
        df[new_col_name]      = df[new_col_name] * df[interactions[2]]
        test_df[new_col_name] = test_df[new_col_name] * test_df[interactions[2]]
        
```

# Creating dummy variables

```python
def createDummies(df_train, df_test, dummy_variables):
    for dummy_var in dummy_variables:
        # getting unique from both datasets
        unique_vals = list(set(df_train[dummy_var].unique()) and set(df_test[dummy_var].unique()))[1:] # Dropping first by default
        for val in unique_vals:
            col_name = "_".join(dummy_var.split()) + "_" + "_".join(str(val).split())
            df_train[col_name] = (df_train[dummy_var].values == val).astype("int")
            df_test[col_name] = (df_test[dummy_var].values == val).astype("int")

        df_train.drop(columns = [dummy_var], inplace = True)
        df_test.drop(columns = [dummy_var], inplace = True)
```

```python
def getBestModel(df, test_df, cv = 5, random_state = 666):
    models = [
        LinearRegression(),
        RidgeCV(),
        LassoCV(cv = cv)
    ]
    
    X_train, X_test, y_train, y_test = train_test_split(df, test_df, random_state = 666)

    # Scaling
    scale = StandardScaler()
    X_train_sc = scale.fit_transform(X_train)
    X_test_sc = scale.transform(X_test)
    
    best_score = 0
    best_model = None
    for model in models:
        model.fit(X_train_sc, y_train)
        score = model.score(X_test_sc, y_test)
        if score > best_score:
            best_score = score
            best_model = model
    print(f"The best score was: {best_score}")
    return (model, scale)
```

### Function to determine best interactions

```python
def _get_combinations(arr,l):
    '''Helper function to returns an array of all 2nd degree combinations from the given array'''
    if len(arr) == l:
        return [arr]
    if l == 1:
        return [[item] for item in arr]
    output = []
    for i in range(2):
        output += [arr[i:i+1] + item  for item in _get_combinations(arr[i+1:], l - 1)]
    return output
```

```python
def get_best_interaction(df, df_test, col_names, target, new_col_name):
    '''
    Given the df DataFrame, returns an interaction Pandas Series from the 
    best combinations of interactions from the given col_names list.
    '''
    best_list = []
    best_corr = 0
    best_col = None
    for i in range(len(col_names)):
        combos = _get_combinations(col_names, i + 1)
        for combo in combos:
            new_df = df[combo].apply(np.prod, axis = 1)
            corr = np.corrcoef(new_df,target)[1,0]
            if abs(corr) > best_corr:
                best_corr = corr
                best_list = combo
                best_col = new_df
    print(f"Best combination found was: {', '.join(best_list)}")
    print(f"Best correlation found was: {round(best_corr,4)}")
    df[new_col_name]      = best_col
    df_test[new_col_name] = df_test[best_list].apply(np.prod, axis = 1)
```

# Model Prep

It would be helpful to have a quick way of assessing models, so I'll make a basic function to do just that.  

Then, I will use the model with the best performance to submit my results with.

```python
from sklearn.metrics import mean_squared_error

def assess_model(X, y, model, rand_state = 666):
    '''
    Assesses model with given variable dataframe X and target y.
    NOTE: assuming input target y is a log transformed value. Also,
          assuming the input data are in need of scaling for model.
    '''
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = rand_state)
    
    # Scaling variables for regularization
    ss = StandardScaler()
    X_train_sc = ss.fit_transform(X_train)
    X_test_sc  = ss.transform(X_test)
    
    # Fitting model
    model.fit(X_train_sc, y_train)
    
    # R^2 score on training set
    r2_train = model.score(X_trian_sc, y_train)
    # R^2 score on testing set
    r2_test = model.score(X_test_sc, y_test)

    # R^2 score on Cross-Validation
    r2_cv = cross_val_score(model, X, y, cv = 5)
    
    # RMSE scores
    rmse_train = np.exp(model.predict(X_train_sc))
    rmse_test  = np.exp(model.predict(X_test_sc))

    print("Train RMSE Score:",  np.sqrt(mean_squared_error(rmse_train, np.exp(y_train))))
    print("Test RMSE Score:",  np.sqrt(mean_squared_error(rmse_test, np.exp(y_test))))
    
    return {
        "r2_train": r2_train,
        "r2_test" : r2_test,
        "r2_cv"   : r2_cv
    }
```
